## Webscapping Loksabha Data 

```python
## Libraries for scrapping 
import requests
import pandas as pd
import time
import json
```

```python
## To store the arrays
loksabha_nums = []
session_nums = []
api_data =[]
api_urls =[]
```

```python
#https://sansad.in/api_ls/member/getMemberAttendanceMonthWise?loksabha=18&session=5&locale=en
BASE_URL = 'https://sansad.in/api_ls/member'
# We found that loksabha start with 13
Loksabha_start = 15 # here we start from 15
# Till now maximum session appear as 18 in loksabha 16

## generating the base url for fetching the api uptil 18 ( As per 2025 only 18 lokhsabha has been there)
loksabha_no = Loksabha_start;
while loksabha_no < 18:
    session_no=1
    while session_no <= 18:
        session_url =  f"{BASE_URL}/getMemberAttendanceMonthWise?loksabha={loksabha_no}&session={session_no}&locale=en"
        
        print(session_url)
        try:
            session_response = requests.get(session_url)
            session_response.raise_for_status() # to check the error 
            sessions = session_response.json()
            #Condition when that session is not there ( < 18 session in any lok sabha ) 
            if len(sessions) == 0 : # if return nothing
                break

            # Adding the datas inside the array
            loksabha_nums.append(loksabha_no)
            session_nums.append(session_no)
            api_urls.append(session_url)
            api_data.append(sessions)
            session_no += 1

        except requests.exceptions.RequestException as e:
            print({e})
    loksabha_no += 1
time.sleep(1) # stop for 1 s to move smoothly        

```

```python
# Creating a Data Frame
Loksabha_df = pd.DataFrame(
    {
        'loksabha_no.': loksabha_nums,
        'sessions_no.': session_nums,
        'api_data': api_data,
        'api_url':api_urls
        
    }
)

```

```python
Loksabha_df.head()
Loksabha_df['api_data'][1]
```

```python
## As all of this comes under the the same column without grouping so we explode the api_data column
Loksabha_df_exploded= Loksabha_df.explode('api_data')
Loksabha_df_exploded['api_data'][1]
```

```python
#reset the index and drop the column index data
Loksabha_df_exploded= Loksabha_df.explode('api_data').reset_index().drop(columns='index')
Loksabha_df_exploded
```

```python
## To extarct the keys of the api_data we use lambda function
Loksabha_df_exploded['dateOfAttendance']=Loksabha_df_exploded['api_data'].apply(lambda x : x['dateOfAttendance'])
Loksabha_df_exploded['dateOfAttendance']=pd.to_datetime(Loksabha_df_exploded['dateOfAttendance'])
Loksabha_df_exploded['dateOfAttendance']=Loksabha_df_exploded['dateOfAttendance'].dt.strftime('%Y/%m/%d')
```

#### Now have to get the member attendence from that api in the same fashion

```python
ls_member_df = Loksabha_df_exploded.copy()
ls_member_df
```

```python
# store list
attendence_list = []

# https://sansad.in/api_ls/member/getMemberAttendanceDateWise?loksabha=16&session=8&dateOfAttendance=2016/05/11&locale=en
for index,row in ls_member_df.iterrows():
    loksabha_no = row['loksabha_no.']
    session_no = row['sessions_no.']
    attendence_date = row['dateOfAttendance']
    session_url =  f"{BASE_URL}/getMemberAttendanceDateWise?loksabha={loksabha_no}&session={session_no}&dateOfAttendance={attendence_date}&locale=en"
    print(session_url)
    session_response = requests.get(session_url)
    sessions = session_response.json()
    attendence_list.append(sessions)

    #time.sleep(1)
```

```python
ls_member_df['member_attendence'] = attendence_list
ls_member_df
```

```python
# As all the cell in member_attendence column has every name we have to sue other columns and make them there for every single loksabha and session and date according to their name
```

```python
# Export to excel
ls_member_df.to_csv('member_attendance.csv')
```
